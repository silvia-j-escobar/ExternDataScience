```python
import json

# Load JSON file
with open('sample.json', 'r') as file:
    data = json.load(file)

# Display the loaded data
print(data)
```

```python
# Retrieve all order IDs and total amounts
order_details = []

# Iterate through each customer
for customer in data:
  # Iterate through each order for the current customer
  if 'orders' in customer:
    for order in customer['orders']:
      # Extract order_id and total and add to the list
      order_details.append({
          'order_id': order.get('order_id'),
          'total': order.get('total')
      })

# Display the extracted order details
print(order_details)
```

```python
# Code thar normalizes the JSON structure using Pandas
import pandas as pd

# Normalize the JSON data
normalized_data = []
for customer in data:
    customer_id = customer.get('customer_id')
    for order in customer.get('orders', []):
        order_id = order.get('order_id')
        order_total = order.get('total')
        for item in order.get('items', []):
            normalized_data.append({
                'customer_id': customer_id,
                'order_id': order_id,
                'product': item.get('product'),
                'quantity': item.get('quantity'),
                'total': order_total  # Using the order total as requested
            })

# Create a DataFrame
df = pd.DataFrame(normalized_data)

# Display the DataFrame
display(df)
```

```python
# Save the DataFrame to a JSON file
output_filename = 'normalized_data.json'
df.to_json(output_filename, orient='records', indent=2)

# Provide a link to download the file
from google.colab import files
files.download(output_filename)
```
