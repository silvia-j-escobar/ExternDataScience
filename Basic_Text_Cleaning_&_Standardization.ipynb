```python
text = "  This   is  a   test   text   with    extra   spaces.   \n\n"
```

```python
import re
clean_text = re.sub(r'\s+', ' ', text).strip()
print(clean_text)  # "This is a test text with extra spaces."
```

```python
import unicodedata
import ftfy

text = "Th√≠s √≠s √† pr√≥blem text"
fixed_text = ftfy.fix_text(text)  # Fix corrupted text
normalized_text = unicodedata.normalize('NFKD', fixed_text).encode('ascii', 'ignore').decode()
print(normalized_text)  # "This is a problem text"
```
